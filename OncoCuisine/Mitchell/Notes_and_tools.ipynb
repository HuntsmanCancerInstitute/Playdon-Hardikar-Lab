```python
import os
from bs4 import BeautifulSoup

# Folder containing your HTML files
folder_path = '/Users/mitch/OneDrive/Desktop/Mitch_hell_CAP/OncoCuisine_AI/food_recipes_havard'

#'/Users/mitch/OneDrive/Desktop/Mitch_hell_CAP/OncoCuisine_AI/food_recipes_havard/food_recipes_havard'

# Loop through all HTML files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.html'):
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
            soup = BeautifulSoup(file, 'html.parser')
            
            # Extract title (commonly in <title> or <h1>)
            title = soup.find('h1')
            if not title:
                title = soup.find('title')
            title_text = title.get_text(strip=True) if title else 'No title found'
            
            # Extract ingredients (example: <ul class="ingredientsList"> or <div class="ingredient-description">)
            ingredients = []
            for ul in soup.find_all('ul', class_='ingredientsList'):
                for li in ul.find_all('li'):
                    ingredients.append(li.get_text(strip=True))
            if not ingredients:
                for div in soup.find_all('div', class_='ingredient-description'):
                    ingredients.append(div.get_text(strip=True))
            
            # Extract instructions (example: <div class="recipe-method-step-content">)
            instructions = []
            for div in soup.find_all('div', class_='recipe-method-step-content'):
                instructions.append(div.get_text(strip=True))
            
            # Output or save the results
            print(f'File: {filename}')
            print(f'Title: {title_text}')
            print('Ingredients:')
            for ing in ingredients:
                print(f'- {ing}')
            print('Instructions:')
            for step in instructions:
                print(f'- {step}')
            print('\n' + '='*40 + '\n')
```

```python
import requests

  # Replace with your USDA API key

def search_food(query, api_key):
    url = f'https://api.nal.usda.gov/fdc/v1/foods/search?api_key={api_key}&query={query}'
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Search error: {response.status_code}")
        return []
    results = response.json().get('foods', [])
    return [(food['fdcId'], food['description']) for food in results]

def get_nutrient_info(fdc_id, api_key):
    url = f'https://api.nal.usda.gov/fdc/v1/food/{fdc_id}?api_key={api_key}'
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Detail error: {response.status_code}")
        return {}
    data = response.json()
    nutrients = {}
    for n in data.get("foodNutrients", []):
        name = n.get("nutrient", {}).get("name", "")
        amount = n.get("amount", 0)
        unit = n.get("nutrient", {}).get("unitName", "")
        nutrients[name] = f"{amount} {unit}"
    return nutrients

# Example usage:
ingredient = "lettuce"
search_results = search_food(ingredient, api_key)
if search_results:
    fdc_id, description = search_results[0]
    print(f"Top match: {description} (FDC_ID: {fdc_id})")
    nutrients = get_nutrient_info(fdc_id, api_key)
    print("Nutrient breakdown:")
    for k, v in nutrients.items():
        print(f"  {k}: {v}")
else:
    print("No results found.")
```

```python
import fitz  # PyMuPDF
import pandas as pd
import re

#pdf_path = "/Users/mitch/OneDrive/Desktop/Mitch_hell_CAP/OncoCuisine_AI/Recipes/Quiche_Self-Forming_Crust_6_Servings.pdf"
doc = fitz.open(pdf_path)

# Extract all text
all_text = ""
for page in doc:
    all_text += page.get_text()

# Find the Ingredients section
ingredients_section = ""
found_ingredients = False
for line in all_text.splitlines():
    if "Ingredients" in line:
        found_ingredients = True
        continue
    if found_ingredients:
        if line.strip() == "" or "Directions" in line:
            break
        ingredients_section += line + "\n"

# Parse ingredients and quantities
ingredient_list = []
for line in ingredients_section.strip().split("\n"):
    # Basic pattern: quantity at the start (e.g., "2 cups milk")
    match = re.match(r"^([\d\/\.\-\s]+[a-zA-Z]*\.?)\s+(.*)", line)
    if match:
        quantity = match.group(1).strip()
        ingredient = match.group(2).strip()
        ingredient_list.append({"Quantity": quantity, "Ingredient": ingredient})
    else:
        # Fallback: treat the whole line as ingredient if no quantity found
        ingredient_list.append({"Quantity": "", "Ingredient": line.strip()})

# Convert to DataFrame for tabular display
df = pd.DataFrame(ingredient_list)
print(df)
```

```python
import fitz  # PyMuPDF
import re

#pdf_path = "/Users/mitch/OneDrive/Desktop/Mitch_hell_CAP/OncoCuisine_AI/Recipes/Quiche_Self-Forming_Crust_6_Servings.pdf"
doc = fitz.open(pdf_path)

# Extract all text
all_text = ""
for page in doc:
    all_text += page.get_text()

# Find the Ingredients section
ingredients_lines = []
in_ingredients = False
for line in all_text.splitlines():
    if "Ingredients" in line:
        in_ingredients = True
        continue
    if in_ingredients:
        # Stop at the next section header or empty line
        if line.strip() == "" or "Directions" in line or "Instructions" in line:
            break
        ingredients_lines.append(line.strip())

# Clean up each ingredient line: remove leading quantity/unit
ingredient_names = []
for line in ingredients_lines:
    # Remove leading quantity/unit using regex: matches numbers, fractions, and units at the start
    cleaned = re.sub(r"^([\d\s\/\.\-]+[a-zA-Z]*\.?\s*)+", "", line)
    cleaned = cleaned.strip(",.-:; ")  # Remove trailing punctuation
    if cleaned:
        ingredient_names.append(cleaned)

# Output as a Markdown-formatted list
print("## Ingredients\n")
for ingredient in ingredient_names:
    print(f"- {ingredient}")
```
